# Ejercicio aplicado de DataFrames y Spark SQL

```python
#Importando librerías 
import findspark
findspark.init()

import pandas as pd
import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local[*]")\
        .appName('PySpark_Df')\
        .getOrCreate()
```

```python
#Importando el dataframe Word Cup Players csv
fifa_df = spark.read.csv ("WorldCupPlayers.csv", inferSchema = True, header = True)
fifa_df.show()
```

```python
#Visualizando el tipo de variables
fifa_df.printSchema()
```

```python
#Contando elementos
fifa_df.count()
```

```python
#Extrayendo valor estadisticos de la columna posicion
fifa_df.describe('Position').show()
```

```python
#Extrayendo nombres de jugares y entrenadores asociados
fifa_df.select('Player Name','Coach Name').distinct().show()
```

```python
#Número de partidos con el ID 1096 existentes en el dataframe
fifa_df.filter(fifa_df.MatchID =='1096').count()
```

```python
#Filtrando datos cuya posicion y evento coincidan con el criterio
fifa_df.filter((fifa_df.Position == 'C') & (fifa_df.Event=="G40'")).show()
```

```python
#Mostrando registros donde el MatchID es >= 20
fifa_df.createOrReplaceTempView("temp_table")

spark.sql("select * from temp_table where MatchID >= 20").show()
```
